```r
# Loading data from anvil
myDF <- read.csv('/anvil/projects/tdm/data/movies_and_tv/rotten_tomatoes_movies.csv')

# Removing unrequired columns
myDF <- myDF[ , !names(myDF) %in% c("cast", "movie_info")]

# Checking size
dim(myDF)

# checking ratings
table(myDF$rating)

# Making simple plot for Ratings
plot(table(myDF$rating), main="Ratings Count", xlab="Rating", ylab="Frequency", col="lightblue")
```

```r
# Finding no. of ratings
length(table(myDF$rating))

# Finding no. of directors
length(table(myDF$directors))

# Finding no.of studios
length(table(myDF$studio_name))

# Finding and sorting top directors so the biggest ones are at the end
tail(sort(table(myDF$directors)))

# Finding biggest studio
tail(sort(table(myDF$studio_name)))
```

```r
# Counting where comedy is only "Comedy"
sum(myDF$genre == "Comedy")

# Using grep to find a mention of Comedy
length(grep("Comedy", myDF$genre))

# Doing the same for Romance
sum(myDF$genre == "Romance")
length(grep("Romance", myDF$genre))
```

```r
# Fixing the typos in the rating column
myDF$rating[myDF$rating == "R)"] <- "R"
myDF$rating[myDF$rating == "PG-13)"] <- "PG-13"

# Making the big table
table_data <- table(myDF$rating, myDF$tomatometer_status)

# Plotting it
mosaicplot(table_data, main="Rating vs Tomatometer", color=c("red", "green", "blue"))
```

```r
# Pasting them together to see the combos
combined_cols <- paste(myDF$rating, myDF$tomatometer_status)

# Seeing which combos are most common
sort(table(combined_cols))

# Grabbing the PG movies that are certified fresh
sub_data <- subset(myDF, rating=="PG" & tomatometer_status=="Certified Fresh")

# Checking the first few
head(sub_data)

# Picking my favorite movies
my_picks <- subset(myDF, movie_title %in% c("Toy Story", "Finding Nemo"))
print(my_picks)
```
